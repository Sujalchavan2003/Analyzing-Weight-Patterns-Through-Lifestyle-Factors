# 1 gender distribution vs weight

This pie chart represents the average weight distribution by gender in the dataset.

The chart shows that males contribute to a slightly higher proportion (52.4%) of the total average weight compared to females (47.6%).

This indicates that, on average, male participants tend to have slightly higher body weight than female participants in this sample.

The difference, though small, may be influenced by factors such as body composition, physical activity levels, or dietary habits.

Both genders, however, have almost balanced representation, suggesting that the dataset is relatively gender-balanced for analysis

```python
avg_weight_by_gender = od.groupby("Gender")["Weight"].mean()
plt.figure(figsize=(6,6))
plt.pie(avg_weight_by_gender, 
        labels=avg_weight_by_gender.index, 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=['lightblue', 'pink'])
plt.title("Average Weight Distribution by Gender")
plt.show()
```

# 4 FCVC: Frequency of consumption of vegetables vs weight

This chart compares average body weight across different levels of vegetable consumption frequency — categorized as Low, Medium, and High.
Individuals with low vegetable consumption show an average weight of around 68 kg.
Those with medium consumption have a slightly higher average weight (~79 kg).
The highest vegetable consumers have an average weight close to 88 kg.
Interestingly, instead of decreasing, weight increases with higher vegetable consumption frequency in this dataset.
This may suggest that:
People who eat more vegetables might also consume more overall calories, or
The relationship between weight and vegetable intake is influenced by other lifestyle factors (like physical activity or total food intake).

```python
od.CALC.value_counts()
```

This box plot compares the distribution of body weight across different levels of alcohol consumption:

No Alcohol:
Individuals who don’t consume alcohol have weights mostly between 60–90 kg, with a few lighter and heavier outliers.

Sometimes:
This group shows the widest weight range, from around 40 kg to over 160 kg, indicating high variability in body weights among occasional drinkers.

Frequently:
Regular drinkers have a narrower range, with weights mostly between 70–95 kg, suggesting more consistent body weight patterns.

Always:
The few individuals who always consume alcohol show lower and stable weights, around 65 kg, with no noticeable variation

# 6 MTRANS comaprison with weight

```python
od.MTRANS.value_counts()
```

This bar plot titled *“Daily Transportation Affect Weight”* illustrates how different modes of daily transportation are associated with average body weight (in kilograms). Here’s a breakdown of what the chart shows:

* *X-axis:* Represents modes of daily transportation — Public Transportation, Walking, Automobile, Motorbike, and Bike.
* *Y-axis:* Indicates average *weight in kilograms (KGs).*
* *Error bars:* Show variability (like standard deviation or confidence intervals) for each transportation mode’s average weight.

### Interpretation:

1. *Public Transportation users* have the *highest average weight, around **85–90 kg*.

   * This could suggest less physical activity compared to walking or biking.

2. *Walking commuters* have a *lower average weight (~70 kg)*, which makes sense as walking increases daily physical activity.

3. *Automobile users* have a *high average weight (~85 kg)*, similar to public transport users — possibly due to sedentary commuting habits.

4. *Motorbike users* fall in the *middle (~75 kg)* range.

5. *Bike commuters* have a *moderate-to-high weight (~78 kg), though the **error bar is large*, indicating high variation — some may be quite fit, while others not as much.

### Overall Insight:

* People who engage in *active transportation* (like walking or biking) tend to have *lower or more variable body weight*.
* *Sedentary commuting modes* (public transport, automobile) are linked with *higher average weight*.

This plot titled *“Obesity Levels Distribution”* is a *horizontal bar chart* that shows how many people fall into different obesity or weight categories.

Here’s how to interpret it step by step:

* *X-axis (horizontal):* Represents the *count* (number of people in each category).
* *Y-axis (vertical):* Represents the *obesity levels, ranging from **Insufficient Weight* to *Obesity Type III*.

1. *Insufficient_Weight* – People who are underweight.
2. *Normal_Weight* – People with a healthy weight.
3. *Overweight_Level_I*
4. *Overweight_Level_II*
5. *Obesity_Type_I*
6. *Obesity_Type_II*
7. Obesity_Type_III

Obesity_Type_I* has the *highest count*, meaning most individuals in the dataset fall under this category.
Obesity_Type_III* and *Obesity_Type_II* also have relatively high counts, showing a significant portion of the population is obese.
Normal_Weight and Insufficient_Weight* have *fewer counts*, indicating fewer people have healthy or low body weight compared to overweight or obese categories.
The data suggests that *obesity is quite prevalent* in this dataset — especially *Obesity Type I* being the most common.
This could indicate lifestyle or dietary trends leading to higher obesity rates among the surveyed individuals.

# Overall Insights

Weight seems to be influenced by a combination of lifestyle habits, not by one single behavior.

Moderation appears key — extreme behaviors (always drinking, eating too little or too much) affect body weight patterns.

There’s no simple linear relationship between one lifestyle factor and weight; instead, interactions between habits (diet, alcohol, and gender) play a major role
